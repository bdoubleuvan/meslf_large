# Move files

```python
results_start = 'results_nnz'
results_end = 'results'
file = 'summary_lu.txt'

directories = [p for p in os.walk(os.path.abspath('../../{}'.format(results_start)))]

directories = directories[2][1]

case = 'case9241pegase_GasLib-4197'
slack = 'slack_position_e_0_g_11'

solver = 'gmres'
preconditioner = 'ilu(8)'

p2g_units_max = 3
gfg_units_max = 4

for i, directory in enumerate(directories):
    print(i, directory)

for directory in directories:
    if False: # 'clones_e_16_clones_g_16_clones_c_1_merges_e_64_merges_g_4' == directory:
        pass
    else:
        if solver in ['lu', 'direct']:
            for p2g in range(p2g_units_max+1):
                for gfg in range(gfg_units_max+1):
                    start_path = os.path.join(os.path.abspath('../../{}/{}/{}'.format(results_start, case, slack)),
                                              directory,
                                              "p2g_{}_gfg_{}".format(p2g, gfg),
                                              solver,
                                              file)
                    end_path = os.path.join('D:{}/{}/{}'.format(results_end, case, slack),
                                            directory,
                                            "p2g_{}_gfg_{}".format(p2g, gfg),
                                            solver)
                    if os.path.exists(start_path):
                        print(start_path)
                        os.makedirs(end_path, exist_ok=True)
                        shutil.move(start_path, end_path)
        else:
            for p2g in range(p2g_units_max+1):
                for gfg in range(gfg_units_max+1):
                    start_path = os.path.join(os.path.abspath('../../{}/{}/{}'.format(results_start, case, slack)),
                                              directory,
                                              "p2g_{}_gfg_{}".format(p2g, gfg),
                                              solver,
                                              preconditioner,
                                              file)
                    end_path = os.path.join('D:{}/{}/{}'.format(results_end, case, slack),
                                            directory,
                                            "p2g_{}_gfg_{}".format(p2g, gfg),
                                            solver,
                                            preconditioner)
                    if os.path.exists(start_path):
                        print(start_path)
                        os.makedirs(end_path, exist_ok=True)
                        shutil.move(start_path, end_path)
```
